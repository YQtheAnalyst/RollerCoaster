### Step 5: Ask a Question

```python
df.head(5)
```

- What are the locations with the fastest roller coaster (minimum of 10)?

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind = 'barh', title = 'Average Coast Speed by Location')
ax.set_xlable = 'Average Coaster Speed'
plt.show()
```

- Which roller coasters in operation has the highest speed?

```python
ax = df[['Coaster_Name', 'Speed_mph']] \
    .sort_values(by = 'Speed_mph', ascending = False) \
    .head(10) \
    .set_index('Coaster_Name') \
    .plot(kind='barh')

ax.invert_yaxis()
ax.set_xlabel('Speed_mph')
ax.set_ylabel('')
ax.set_title('Top 10 Fastest Roller Coasters')

plt.show()
```
